## Tercera Parte (20 puntos)

```python
from transformers import AutoTokenizer
from datasets import load_dataset

tokenizer = AutoTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-cased")

dataset = load_dataset('Biobert_json.py', data_dir='./data56/', trust_remote_code=True)
```

#### Punto 3.1: Desplegar los tres conjuntos del dataset con sus respectivas características `sentencia` y `tag`

```python
train_dataset = dataset['train']
validation_dataset = dataset['validation']
test_dataset = dataset['test']

print("Conjunto de entrenamiento:", train_dataset)
print("Conjunto de validación:", validation_dataset)
print("Conjunto de test:", test_dataset)
```

#### Punto 3.2: Desplegar todas las etiquetas `tag` del conjunto de entrenamiento y de prueba


```python
train_tags = [element['tag'] for element in train_dataset]
print("Etiquetas del conjunto de entrenamiento:", train_tags)

test_tags = [element['tag'] for element in test_dataset]
print("Etiquetas del conjunto de test:", test_tags)  
```

##### Mapeo de etiquetas: Mostrar los nombres correspondientes a cada número

```python
label_names = [
    "B_CANCER_CONCEPT", "B_CHEMOTHERAPY", "B_DATE", "B_DRUG", "B_FAMILY", 
    "B_FREQ", "B_IMPLICIT_DATE", "B_INTERVAL", "B_METRIC", "B_OCURRENCE_EVENT", 
    "B_QUANTITY", "B_RADIOTHERAPY", "B_SMOKER_STATUS", "B_STAGE", "B_SURGERY", 
    "B_TNM", "I_CANCER_CONCEPT", "I_DATE", "I_DRUG", "I_FAMILY", "I_FREQ", 
    "I_IMPLICIT_DATE", "I_INTERVAL", "I_METRIC", "I_OCURRENCE_EVENT", "I_SMOKER_STATUS", 
    "I_STAGE", "I_SURGERY", "I_TNM", "O"
]

mapped_train_tags = [[label_names[tag] for tag in tags] for tags in train_tags[:5]]
print("Etiquetas de entrenamiento con nombres (primeras 5):")
for tags in mapped_train_tags:
    print(tags)

mapped_test_tags = [[label_names[tag] for tag in tags] for tags in test_tags[:5]]
print("\nEtiquetas de test con nombres (primeras 5):")
for tags in mapped_test_tags:
    print(tags)
```

#### Punto 3.3: Implementar código para desplegar una sentencia de palabras `train_{0}` con sus respectivas etiquetas

```python
example_sentence = train_dataset[0]['sentencia']
example_tags = train_dataset[0]['tag']

print("Sentencia:", example_sentence)
print("Etiquetas:", example_tags)
mapped_test_tags = [label_names[tag] for tag in example_tags]
print("\nEtiquetas con nombres:", mapped_test_tags)
```

#### Punto 3.4: Tokenizar una sentencia del dataset con el tokenizador BETO

```python
tokenized_sentence = tokenizer.tokenize(' '.join(example_sentence))

print("Sentencia tokenizada:", tokenized_sentence)
```
